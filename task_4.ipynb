```python
# from libs import normalization, equalization, filters, noise, graphs, edge_detection, threshold
from libs import colorMap,segmentation, thresholding2
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("images/lenna.png")
img = cv2.cvtColor(img,cv2.COLOR_BGR2GRAY)
global_optimal = thresholding2.optimal(img)
local_optimal = thresholding2.LocalThresholding(img , 4 , thresholding2.optimal)
fig, axes = plt.subplots(1, 3, figsize=(15, 6))
ax = axes.ravel()

ax[0].imshow(img , cmap="gray")
ax[0].set_title('Original Image')
ax[0].set_axis_off()

ax[1].imshow(global_optimal , cmap="gray" )
ax[1].set_title('Global_Optimal')
ax[1].set_axis_off()


ax[2].imshow(local_optimal , cmap="gray"  )
ax[2].set_title('Local_Optimal')
ax[2].set_axis_off()
plt.tight_layout()
plt.show()
```
