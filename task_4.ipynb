```python
# from libs import normalization, equalization, filters, noise, graphs, edge_detection, threshold
from libs import colorMap,segmentation
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

```python
img = cv2.imread("images/deer.png")
b,g,r = cv2.split(img)       # get b,g,r
img = cv2.merge([r,g,b])     # switch it to rgb
cop_img = np.copy(img)
luv_img = colorMap.RGB2LUV(img)
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
ax = axes.ravel()

ax[0].imshow(cop_img)
ax[0].set_title('Original Image')
ax[0].set_axis_off()

ax[1].imshow(luv_img )
ax[1].set_title('Luv Image')
ax[1].axis('image')

plt.tight_layout()
plt.show()
```

```python
img = cv2.imread("images/test.png")
cop_img = np.copy(img)
segmented_image ,labels = segmentation.apply_k_means (img)
fig, axes = plt.subplots(1, 2, figsize=(15, 6))
ax = axes.ravel()

ax[0].imshow(cop_img)
ax[0].set_title('Original Image')
ax[0].set_axis_off()

ax[1].imshow(segmented_image )
ax[1].set_title('Luv Image')
ax[1].axis('image')

plt.tight_layout()
plt.show()
```
