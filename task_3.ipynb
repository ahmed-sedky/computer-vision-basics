**Import numpy and the required image processing functions**

*Note: cv2 is only imported to open the image and get its dimensions*

```python
# from libs import normalization, equalization, filters, noise, graphs, edge_detection, threshold
from libs import filters, edge_detection ,harris
import cv2
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image
```

**Open the image**

```python
img = cv2.imread("images/squares.png")

copImg = np.copy(img)
harriResponse  = harris.harrisCorner(img,0.04)
cornerImage = harris.corner2Image(img, harriResponse , 0.05)

fig, axes = plt.subplots(1, 3, figsize=(15, 6))
ax = axes.ravel()

ax[0].imshow(copImg)
ax[0].set_title('Input image')
ax[0].set_axis_off()

ax[1].imshow(harriResponse ,cmap = "gray")
ax[1].set_title('corner_img')
ax[1].axis('image')

ax[2].imshow(cornerImage)
ax[2].set_title('Output image')
ax[2].set_axis_off()


plt.tight_layout()
plt.show()
```
